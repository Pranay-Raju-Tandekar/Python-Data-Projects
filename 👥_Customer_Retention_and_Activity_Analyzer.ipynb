# 👥 Customer Retention & Activity Analyzer
An automated CRM analytics engine built using **Python**, **NumPy**, and **Pandas** to process user interaction metrics, identify churn risks, and isolate high-engagement customer accounts.

```python
import numpy as np
import pandas as pd

# 1. Custom function to evaluate user engagement status
def categorize_engagement(days_active):
    """Flags accounts based on their active usage patterns."""
    # If active less than 3 days a week, flag as a retention risk
    return np.where(days_active >= 3, "Active / Safe", "Retention Risk")

print("🚀 Initializing CRM Analytics Data Stream...")

# 2. Raw Customer Interaction Logs
crm_data = {
    "Customer_ID": [501, 502, 503, 504, 505, 506],
    "Username": ["Aarav_D", "Isha_K", "Kabir_M", "Riya_S", "Zane_Tech", "Meera_R"],
    "Weekly_Days_Active": [5, 1, 4, 2, 7, 0],
    "Monthly_Spend_INR": [4200, 850, 6100, 0, 12500, 3100]
}

# Convert dictionary into a structured CRM DataFrame
customer_df = pd.DataFrame(crm_data)

# 3. Apply operational flags using our vectorized status rule
customer_df["Account_Status"] = categorize_engagement(customer_df["Weekly_Days_Active"])

print("\n============ CRM CUSTOMER INSIGHTS DASHBOARD ============")
print(customer_df)
print("=========================================================")

# 4. Filter for high-risk accounts that require immediate marketing outreach
risk_mask = customer_df["Account_Status"] == "Retention Risk"
risk_accounts_df = customer_df[risk_mask]

print("⚠️ CHURN RISK WARNING - IMMEDIATE OUTREACH REQUIRED:")
print(risk_accounts_df[["Username", "Weekly_Days_Active", "Monthly_Spend_INR"]])
print("=========================================================")
```
