# 📊 Advanced E-Commerce Sales Analytics Dashboard
Developed a high-performance transactional data pipeline using **Python**, **NumPy**, and **Pandas** to process, analyze, and audit corporate sales records.

### 🚀 Key Engineering Implementations:
* **Functional Modularity:** Implemented reusable calculation modules with default parameter logic to dynamically isolate regional tax overhead (VAT).
* **Vectorized Data Architecture:** Eliminated traditional, resource-heavy Python `for` loops by leveraging **NumPy vectorization** to calculate gross and total product revenue instantaneously across massive arrays.
* **Boolean Masking Filters:** Engineered high-speed data stencils to dynamically slice datasets, isolating transactions generating greater than Rs. 50,000 for executive reporting.
* **Automated Data Export:** Integrated specialized Pandas operational pipelines to instantly format and export cleaned, targeted datasets into industry-standard `.csv` spreadsheets for business intelligence use.

---

```python
import numpy as np
import pandas as pd

# 1. Reusable tax calculation module (Level 3 Functions)
def calculate_vat(price_column, tax_rate=0.18):
    """Calculates a standard 18% tax on sales data."""
    return price_column * tax_rate

print("⚙️ Initializing E-Commerce Data Pipeline...")

# 2. Raw Transaction Database (Level 5 Pandas DataFrames)
sales_records = {
    "Transaction_ID": [1001, 1002, 1003, 1004, 1005],
    "Product_Name": ["Wireless Mouse", "Gaming Laptop", "Mechanical Keyboard", "4K Monitor", "USB-C Hub"],
    "Base_Price": [1500, 75000, 4500, 28000, 2500],
    "Units_Sold": [12, 3, 8, 5, 25]
}

df = pd.DataFrame(sales_records)

# 3. Vectorized Financial Math (Level 4 NumPy Calculations)
# Calculate gross revenue per product instantly without loops
df["Gross_Revenue"] = df["Base_Price"] * df["Units_Sold"]

# Inject calculated tax tracking using our function
df["Tax_Amount"] = calculate_vat(df["Gross_Revenue"])

# Calculate final total revenue
df["Total_Revenue"] = df["Gross_Revenue"] + df["Tax_Amount"]

print("\n============ ENTERPRISE SALES ANALYTICS ============")
print(df)
print("====================================================")

# 4. Descriptive Statistics & Insights
total_enterprise_sales = df["Total_Revenue"].sum()
average_order_value = df["Base_Price"].mean()

print(f"Total Portfolio Revenue Generates : Rs. {total_enterprise_sales:,}")
print(f"Average Product Catalog Value    : Rs. {average_order_value:,.2f}")
print("====================================================")

# 1. Create a Boolean Mask to find rows where Total_Revenue is greater than Rs. 50,000
high_value_mask = df["Total_Revenue"] > 50000

print("============ HIGH-VALUE TRANSACTION REPORT ============")
print("Step 1: The Underlying Boolean Mask Array:")
print(high_value_mask.values)
print("-------------------------------------------------------")

# 2. Pass the mask into the DataFrame to slice out the actual matching rows
high_value_df = df[high_value_mask]

print("Step 2: Filtered High-Revenue Transactions:")
print(high_value_df[["Product_Name", "Units_Sold", "Total_Revenue"]])
print("=======================================================")


# 1. Export the high-value transaction DataFrame to a CSV file
# index=False ensures Pandas doesn't add an extra column of row numbers (0, 1, 2) to your file
high_value_df.to_csv("High_Value_Sales_Report.csv", index=False)

print("✅ Success! 'High_Value_Sales_Report.csv' has been generated in your Colab file system.")
```

```python
import matplotlib.pyplot as plt

print("📊 Generating Visual Analytics...")

# 1. Choose the data columns to plot
# We will put Product Names on the X-axis and Total Revenue on the Y-axis
products = df["Product_Name"]
revenue = df["Total_Revenue"]

# 2. Create a clean bar chart
plt.figure(figsize=(8, 5))  # Sets the size of your graph window
plt.bar(products, revenue, color='skyblue', edgecolor='royalblue')

# 3. Add styling titles and labels
plt.title("Total Revenue Generated per Product", fontsize=14, fontweight='bold')
plt.xlabel("Product Name", fontsize=12)
plt.ylabel("Revenue (in INR)", fontsize=12)

# 4. Clean up the layout and display the graph
plt.xticks(rotation=15)  # Rotates product names slightly so they don't overlap
plt.tight_layout()
plt.show()
```
