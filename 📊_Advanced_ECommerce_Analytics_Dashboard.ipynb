🏗️ Building the E-Commerce Analytics Engine

```python
import numpy as np
import pandas as pd

# 1. Reusable tax calculation module (Level 3 Functions)
def calculate_vat(price_column, tax_rate=0.18):
    """Calculates a standard 18% tax on sales data."""
    return price_column * tax_rate

print("⚙️ Initializing E-Commerce Data Pipeline...")

# 2. Raw Transaction Database (Level 5 Pandas DataFrames)
sales_records = {
    "Transaction_ID": [1001, 1002, 1003, 1004, 1005],
    "Product_Name": ["Wireless Mouse", "Gaming Laptop", "Mechanical Keyboard", "4K Monitor", "USB-C Hub"],
    "Base_Price": [1500, 75000, 4500, 28000, 2500],
    "Units_Sold": [12, 3, 8, 5, 25]
}

df = pd.DataFrame(sales_records)

# 3. Vectorized Financial Math (Level 4 NumPy Calculations)
# Calculate gross revenue per product instantly without loops
df["Gross_Revenue"] = df["Base_Price"] * df["Units_Sold"]

# Inject calculated tax tracking using our function
df["Tax_Amount"] = calculate_vat(df["Gross_Revenue"])

# Calculate final total revenue
df["Total_Revenue"] = df["Gross_Revenue"] + df["Tax_Amount"]

print("\n============ ENTERPRISE SALES ANALYTICS ============")
print(df)
print("====================================================")

# 4. Descriptive Statistics & Insights
total_enterprise_sales = df["Total_Revenue"].sum()
average_order_value = df["Base_Price"].mean()

print(f"Total Portfolio Revenue Generates : Rs. {total_enterprise_sales:,}")
print(f"Average Product Catalog Value    : Rs. {average_order_value:,.2f}")
print("====================================================")
```
